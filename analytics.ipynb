```python
# Get data file names
path = r'data/960/'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
winners = []
for key, row in big_frame.iterrows():
    if row.white == 1:
        winners.append(row.white_agent)
    if row.black == 1:
        winners.append(row.black_agent)
winners
big_frame['winner'] = winners
```
